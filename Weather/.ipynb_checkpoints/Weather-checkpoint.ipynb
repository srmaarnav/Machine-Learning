```python
#precip has negative impact on temp
#temp max has big positive input
#min input has a lower impact
```

```python
#creating a function to make predictions
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"2020-12-30"]
    test = core_weather.loc["2021-01-01":]
    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])
    error = mean_absolute_error(test["target"], predictions)
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error,combined
```

```python
#run the function to iterate the predictions and check error
```

```python
core_weather["month_max"] = core_weather["temp_max"].rolling(30).mean()
##rolling mean function creates mean for 30 days in sequence and assign it to month max
```

```python
core_weather
##for  months with no consecutive 30 days before it we get NAN
```

```python
core_weather["month_day_max"] = core_weather["month_max"]/core_weather["temp_max"]
#ratio of monthly max to daily max
```

```python
core_weather
```

```python
core_weather["max_min"] = core_weather["temp_max"]/core_weather["temp_min"]
```

```python
## to add predictors
predictors = ["precip", "temp_max", "temp_min", "month_day_max", "max_min"]
```

```python
core_weather["monthly_avg"] = core_weather["temp_max"].groupby(core_weather.index.month).apply(lambda x: x.expanding(1).mean())
##groups temp max by months, which is easy due to converting the index to date 
##we use lambda expanding function that only includes previous value to generate the mean, otherwise we are in risk of using future values 
##to calculate the mea
##for the last column, first column is itself
##second column mean of first two and so on
```

```python
core_weather
```

```python
##at last the avg is of all the columns
```

```python
core_weather["day_of_year"] = core_weather["temp_max"].groupby(core_weather.index.day_of_year).apply(lambda x: x.expanding(1).mean())
##day of year give whish day is it in that year
```

```python
predictors = ["precip", "temp_max", "temp_min", "month_day_max", "max_min", "monthly_avg", "day_of_year"]
error, combined = create_predictions(predictors, core_weather, reg)
```
